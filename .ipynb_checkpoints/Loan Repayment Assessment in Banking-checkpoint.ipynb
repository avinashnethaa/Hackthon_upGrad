```python
# Display the missing value.
Missing_value = train_df.isnull().sum()*100/len(train_df)
Missingv_value_df = pd.DataFrame({'column_name':train_df.columns,'Percentage_of_missing':Missing_value})
```

```python
# checking the mode.
train_df['emp_length'].mode()
```

```python
print(train_df['pub_rec_bankruptcies'].median())
print(train_df['revol_util'].median())
```

```python
# 1. Impute missing values in emp_length with the mode
train_df['emp_length'].fillna(train_df['emp_length'].mode()[0], inplace=True)

# 3. Impute missing values in numeric columns with median
numeric_columns = ['pub_rec_bankruptcies', 'revol_util','num_actv_bc_tl','mort_acc','tot_cur_bal']
for col in numeric_columns:
    train_df[col].fillna(train_df[col].median(), inplace=True)

# 2 and 6. Considering dropping emp_title and title columns
train_df.drop(['emp_title', 'title'], axis=1, inplace=True)
```

### Observations:
    
<b>emp_length:</b> Missing values in the emp_length column have been imputed with the mode (most frequent value) of the column, which is "10+ years".

<b>pub_rec_bankruptcies and revol_util:</b> Missing values in these numeric columns have been imputed with the median values of their respective columns.

<b>emp_title and title:</b> These columns have been dropped from the DataFrame as per the treatment plan. Since they had a relatively high percentage of missing values and

Before proceeding into visualization,going to convert the target variable from categorical to numerical.

```python
from sklearn.preprocessing import LabelEncoder
# initialize  LabelEncoder
le = LabelEncoder()
# Convert Categorical variable into Numerical labels
train_df['loan_status']= le.fit_transform(train_df['loan_status'])

# print the target varibale
print(train_df['loan_status'].value_counts())
```

Observation: Performed labelencoding to convert the categorical into numerical
    
    Paid : 1
    Defaulted : 0

```python
sns.countplot(train_df,x=train_df['loan_status'])
plt.title('Loan Status Distribution')
plt.show()
```
